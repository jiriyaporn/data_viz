```
import pandas as pd
```

```
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv ')
df
```

### Paranel Codinate

```
pd.plotting.parallel_coordinates(df, 'Name' );
```

```
pd.plotting.parallel_coordinates(df, 'Name' , color=( '#D00000', '#006400', '#FF1493'))## himl colur 
```

```
pd.plotting.parallel_coordinates(df, 'Name' , color=( 'r', 'g', 'b'))
```

```
pd.plotting.parallel_coordinates(df, 'Name' , color=( (1,0,0), (0,0.7,0), (0.6,0,0)))
```

## ปรับ colormap

```
pd.plotting.parallel_coordinates(df, 'Name' , colormap='cool')
```

```
pd.plotting.parallel_coordinates(df, 'Name' , colormap='hsv')
```

```
pd.plotting.parallel_coordinates(df, 'Name' , colormap='tab10')
```

## เริ่ม matplotlip


```
from matplotlib import pyplot as plt ## ทำงานกับ numpy ได้ดี
```

### แปลงค่า data จาก numpy dataframe เป็น numpy array

```
np_data = df.iloc[:,:-1].to_numpy()
np_data
```

## visualize array data pixcel

```
import matplotlib
```

```
matplotlib.rcParams[ 'figure.figsize' ] = [15,15] ## กำหนดขนาดของฟิกเกอรP
```

```
plt.imshow(np_data[:50,:], interpolation= 'nearest' )
```

```
plt.imshow(np_data[:1,:], interpolation= 'nearest' )
```

##การใช้ subplot

```
plt.subplot(1,2,1)
plt.imshow(np_data[:50,:],interpolation='nearest', cmap='magma')

plt.subplot(1,2,2)
plt.imshow(np_data[50:100,:],interpolation='nearest',cmap='magma')

plt.subplot(1,3,3)
plt.imshow(np_data[100:150,:],interpolation='nearest',cmap='magma')
```
